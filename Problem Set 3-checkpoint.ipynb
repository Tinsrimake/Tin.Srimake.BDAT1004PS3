```python
print('step 9')
def stats(loc):
    loc = pd.Series(loc)
    Min = loc.min()
    Max = loc.max()
    Mean = loc.mean()
    Std = loc.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = data.apply(stats)
loc_stats
```

```python
print('step 10')
day_stats = data.apply(stats,axis=1)
day_stats.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
year = data['Year'].values
marriagescapita = data['Marriages_per_1000'].values
divorcescapita = data['Divorces_per_1000'].values

plt.plot(year, marriagescapita)
plt.plot(year, divorcescapita)
plt.title('Marriages and Devorces vs Years')
plt.xlabel('Year')
plt.ylabel('Number of Cases Per Capita')
plt.legend(['Marriages', 'Divorces'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data = data[data['Year'].apply(lambda x: x in [1900, 1950, 2000])]

year = data['Year'].values
marriagescapita = data['Marriages_per_1000'].values
divorcescapita = data['Divorces_per_1000'].values

plt.bar(year, marriagescapita)
plt.bar(year, divorcescapita)
plt.title('Marriages and Devorces vs Years')
plt.xlabel('Year')
plt.ylabel('Number of Cases Per Capita')
plt.legend(['Marriages', 'Divorces'])
plt.show()
```
